將iris data載入


```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.externals.six import StringIO   
from sklearn.tree import export_graphviz
import pydotplus
from sklearn.datasets import load_iris
iris = load_iris()
iris_data = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_data.loc[:,"target"] = iris.target
df = iris_data
```

發現有4個參數 花萼長 花萼寬  花瓣長  花瓣寬

```python
df
```

```python
#算出有多少種類
# 0 = Setosa, 1 = Versicolor, 2 = Virginica
df["target"].value_counts()
```

```python
#查看資料有無缺失
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
#將資料分為train test ,並且分成7:3 (test = 0.3) , 將要使用的參數輸入
X_train, X_test, y_train, y_test = train_test_split(df[['sepal length (cm)','petal length (cm)','sepal width (cm)','petal width (cm)']],iris_data[['target']],test_size = 0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier( max_depth = 4)
```

```python
tree.fit(X_train,y_train)
```

```python
#特徵選擇
tree.feature_importances_
```

```python
y_predict = tree.predict(X_test)
y_predict
```

```python
y_test['target'].values
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict)
```
